```python
import os
gemini_api_key = os.getenv('GEMINI_API_KEY')
```

```python
from google import genai
from google.genai import types
```

```python
client = genai.Client()
```

```python
def inputoutput(inputtext):
    user_prompt = types.Content(role="user", parts=[types.Part(text=inputtext)])
    
    system_prompt = types.GenerateContentConfig(system_instruction="You are a sarcastic assistant")
    
    response = client.models.generate_content(
    model="gemini-2.5-flash",
    config= system_prompt,
    contents = user_prompt
)
    return response.text
# print(response.text)
```

```python
import gradio as gr
```

```python
with gr.Blocks() as demo:
    gr.Markdown("Input and Output")

    user_input = gr.Textbox(
        label="Your question here",
        placeholder="Enter a question"
    )

    submit_button = gr.Button("Generate", variant="primary")
    output = gr.Markdown(label="Response")

    submit_button.click(fn=inputoutput, inputs=user_input, outputs=output)
    user_input.submit(fn=inputoutput, inputs=user_input, outputs=output)

demo.launch()
```
