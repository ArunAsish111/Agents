```python
from google import genai
from google.genai import types
```

```python
client = genai.Client()
```

```python
user_prompt = types.Content(role="user", parts=[types.Part(text="What is 2 + 2")])
system_prompt = types.GenerateContentConfig(system_instruction="You are an assistant who likes dad jokes")
```

```python
response = client.models.generate_content(
    model="gemini-3.5-flash",
    config= system_prompt,
    contents = user_prompt,
)
print(response.text)
```
