```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
import os
gemini_api_key=os.getenv("GEMINI_API_KEY")
print(f"API key is {gemini_api_key[0:7]}")
```

```python
from google import genai
from google.genai import types
client = genai.Client()
```

```python
user_prompt=types.Content(role="user", parts=[types.Part(text="Who am I?")])
system_prompt=types.GenerateContentConfig(system_instruction="You are an intelligent agent")
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=user_prompt,
    config=system_prompt
)
print(response.text)
```
